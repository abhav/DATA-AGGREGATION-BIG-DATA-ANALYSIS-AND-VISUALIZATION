### Collect URL from first Query

```python
api_url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
query = "q='border wall'"
page = "&page="
rest = "&begin_date=20190101&api-key=SY8ZVuGSVQxbHrPK7eQYk9d0A819EXOs"
```

```python
pages = math.ceil(count/10)
pages
```

```python
url_lst = []
```

```python
for p in range(1, pages +1):
    url = api_url + query + page + str(p) + rest
    json_data = requests.get(url).json()
    time.sleep(10)
    print(p)
    for i in json_data['response']['docs']:
        weburl = i['web_url']
        url_lst.append(weburl)
```

```python
updated_url_lst = []
```

```python
for i in url_lst:
    if '/video/' not in i:
        updated_url_lst.append(i)
```

```python
len(updated_url_lst)
```

```python
def TxtCreator(href,n):
    page = requests.get(href)
    soup = BeautifulSoup(page.content, 'html.parser')
    time.sleep(3)
    content=""
    try:
        for i in soup.find('section', {"name" : "articleBody"}).find_all('p'):
            content += i.text
        content +="\n"
    except:
        print("Something went wrong for this url:" +href)
    
    a = str(n)

    with open(os.path.join(path, 'DocQuery' + a + '.txt'), 'a+') as temp_file:
        temp_file.write(content)
```

```python
for i in range(0,len(updated_url_lst)):
    TxtCreator(updated_url_lst[i],1)
    print('Complete:'+ str(i))
```
